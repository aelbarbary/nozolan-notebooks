### Step 1: Download `mysite.zip`

Download the starter code from here:

https://github.com/aelbarbary/nozolan-notebooks/blob/main/create-lost-found-app/code/mysite.zip

Tip: If GitHub shows a page, look for a **Download** button (or **Download raw file**).

### Step 2: Upload the ZIP to PythonAnywhere (same as last class)

1. Open **PythonAnywhere → Files**
2. Open your folder: `mysite/` (it’s usually already there)
3. Upload the file you downloaded: `mysite.zip`

### Step 3: Unzip it

1. Go to **PythonAnywhere → Consoles**
2. Start a **Bash** console
3. Run:

```bash
cd ~/mysite
unzip -o mysite.zip
```

This will put the website files into your `mysite/` folder.

### Step 4: Reload + open your website

1. Go to **PythonAnywhere → Web**
2. Click **Reload**
3. Open your website link (example):
   - `https://YOURUSERNAME.pythonanywhere.com/`

If you see the Lost & Found page, you’re ready!
