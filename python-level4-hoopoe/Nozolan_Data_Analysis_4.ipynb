## `1- Animals in the Quran ðŸ•`

`Letâ€™s explore Allahâ€™s wisdom by looking at the animals mentioned in the Quran, using a dataset of the most frequently mentioned animals!`

> #### 1.1 Create the dataset
- First, letâ€™s import the necessary libraries.
- Next, weâ€™ll create a dictionary with our data.
- Finally, weâ€™ll turn that dictionary into a pandas DataFrame.

```python
# import the necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create a dictionary with our data.
animals_data = {
    "Animal": ["Ant", "Bee", "Camel", "Sheep", "Whale", "Dog", "Elephant"],
    "Mentions": [3, 1, 12, 8, 4, 2, 1],
    "Surah Example": ["An-Naml", "An-Nahl", "Al-Ghashiyah", 
                     "Al-An'am", "As-Saffat", "Al-Kahf", "Al-Fil"]
}
```

```python
# turn the dictionary into a pandas DataFrame.
animals_df = pd.DataFrame(animals_data)
```

```python
print(animals_df)
```

> #### 1.2 Turn data into colorful chars
- explore animals mentioned in the Quran and create a colorful bar chart to see which ones are mentioned the most.

```python
# Set the size of the chart (how big the chart will appear on the screen)
plt.figure(figsize=(10, 5))

# Create a bar chart using animal names and how many times they are mentioned in the Quran
# The bars will be in different colors to make the chart look nice and colorful
plt.bar(animals_df["Animal"], animals_df["Mentions"], color=["gold", "yellowgreen", "skyblue", "pink", "teal", "tan", "gray"])

# Add a title at the top of the chart to explain what it's about
plt.title("Animals in the Quran: Allah's Creation Stories", fontsize=14, fontweight="bold")

# Label the x-axis with "Animal Names" to show whatâ€™s being measured (the animals!)
plt.xlabel("Animal Names", fontsize=12)

# Label the y-axis with "Number of Mentions" to show how many times each animal is mentioned in the Quran
plt.ylabel("Number of Mentions", fontsize=12)

# Add dashed lines along the y-axis to make it easier to read the chart and see the numbers
plt.grid(axis="y", linestyle="--")

# Display the chart on the screen
plt.show()
```

## `2- Languages Spoken by Muslims ðŸŒ`

`Letâ€™s explore how Muslims speak many languages but share one faith!`

`EXCITING FACT ðŸŒŸ: "Did you know Muslims speak over 2,000 languages?  --  Letâ€™s visualize the TOP 6 most common`

> #### 2.1 Create the dataset
- Weâ€™ll create a dictionary with our data.
- Then, wâ€™ll turn that dictionary into a pandas DataFrame.

```python
# create a dictionary with our data.
languages_data = {
    "Language": ["Arabic", "Urdu", "Bengali", "Indonesian", "Persian", "Turkish", "Others (2,000+ languages!)"],
    "Speakers (Millions)": [310, 230, 200, 170, 90, 80, 820],
    "Main Regions": ["Middle East, North Africa", "South Asia", "Bangladesh/India", 
                    "Indonesia/Malaysia", "Iran/Afghanistan", "Turkey/Central Asia", 
                    "Global"]
}
```

```python
# turn the dictionary into a pandas DataFrame.
languages_df = pd.DataFrame(languages_data)
```

```python
print(languages_df)
```

> #### 2.2 Exploring with Pie Charts ðŸ•
- FUN FACT: "Pie charts are like pizza slices â€“ which language 'slice' is biggest?"
- Weâ€™re creating a pie chart to show the most spoken languages among Muslims!

```python
# Labels for the pie chart - the different languages weâ€™re looking at
labels = languages_df["Language"]

# Sizes for the pie chart - how many people speak each language (in millions)
sizes = languages_df["Speakers (Millions)"]

# Colors for the pie slices, making it colorful and fun to look at!
colors = ["gold", "skyblue", "limegreen", "tomato", "violet", "orange", "silver"]

# Set the figure size for our pie chart, making it big and easy to see
plt.figure(figsize=(10, 10))

# Create the pie chart using our data: sizes, labels, and the fun colors!
# We also show the percentage of each language and add some fun details like shadow and a title.
# - autopct: stands for "automatic percent" and allows you to display the percentage of each slice.
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90, 
        shadow=True, textprops={"fontsize": 10})

# Add a title to the chart to explain what itâ€™s about
plt.title("Languages Spoken by Muslims Worldwide ðŸŒ\n(1.9 Billion Total)")

# Show the pie chart on the screen
plt.show()
```
