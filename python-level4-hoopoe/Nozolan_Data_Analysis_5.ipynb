## `1- Grouping Zakat Donations 🕌`


`Zakat is a pillar of Islam! Let’s organize donations to help communities in need!` <br>

**Goal:** `Organize Zakat data to find where help is needed most!`

💡 **HOW IT WORKS:** `We’ll group donations by country and sum them up to see totals!`

> #### 1.1 Create the dataset
- First, let’s import the necessary libraries.
- Next, we’ll create a dictionary with our data.
- Finally, we’ll turn that dictionary into a pandas DataFrame.

```python
# import the necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create a dictionary of Zakat donations by country and year.
zakat_data = {
    "Country": ["Yemen", "Somalia", "Bangladesh", "Yemen", "Somalia", "Bangladesh"],
    "Year": [2020, 2020, 2020, 2023, 2023, 2023],
    "Zakat Collected (USD)": [50000, 30000, 80000, 75000, 45000, 120000]
}
```

```python
# convert the dictionary into a pandas DataFrame.
zakat_df = pd.DataFrame(zakat_data)
```

```python
print("🕌 Zakat Donation Dataset:")
print(zakat_df)
```

> #### 1.2 Group and Aggregate by Country
- 🌟 **FUN FACT:** Grouping is like sorting LEGO bricks by color – let’s sort donations by country!
- 💡 **HOW IT WORKS:** groupby("Country") groups rows by country, and .sum() adds donations.

```python
# Group donations by country and calculate totals
grouped = zakat_df.groupby("Country")["Zakat Collected (USD)"].sum().reset_index()
```

**.reset_index():** makes the result easier to read and work with.

```python
# Let's verify what we have got after grouping the data.
print(grouped)
```

```python
# Create a bar chart
plt.figure(figsize=(6, 4))
plt.bar(grouped["Country"], grouped["Zakat Collected (USD)"], color=["red", "green", "blue"])
plt.title("Total Zakat Collected by Country (2020-2023)", fontsize=14)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Zakat (USD)", fontsize=12)
plt.grid(axis="y", linestyle="--")
plt.show()
```

## `2- Correlation – Ramadan Fasting Hours vs. Latitude 🌍`

`Allah created the sun and moon! Let’s explore how latitude affects fasting!`

**Correlation** `means seeing if two things change together!` <br>
`Example: Do taller people have bigger shoes?`

**Goal:** `Does fasting hours correlate with latitude? Let’s find out!`

**🛠️ HOW IT WORKS:** `Latitude measures distance from the equator. Higher values = farther north/south.`

> #### 2.1 Create the dataset
- We’ll create a dictionary with our data.
- Then, w’ll turn that dictionary into a pandas DataFrame.

```python
ramadan_data = {
    "City": ["Makkah", "Jakarta", "Istanbul", "London", "Cape Town"],
    "Fasting Hours": [14, 13, 15, 17, 12],
    "Latitude": [21.4, -6.2, 41.0, 51.5, -33.9]  # Latitude values in degrees
}
```

```python
# convert the dictionary into a pandas DataFrame.
ramadan_df = pd.DataFrame(ramadan_data)
```

```python
print(ramadan_df)
```

> #### 2.2 Visualize with a Scatter Plot
- Scatter plots are like connecting stars – let’s link fasting and latitude!
- **🛠️ HOW IT WORKS:** Each dot represents a city. If dots slope upward, there’s a correlation!

```python
plt.figure(figsize=(7, 5))
plt.scatter(ramadan_df["Latitude"], ramadan_df["Fasting Hours"], s=100, c="purple", alpha=0.8)
plt.title("Fasting Hours vs. Latitude", fontsize=14)
plt.xlabel("Latitude (Degrees)", fontsize=12)
plt.ylabel("Fasting Hours", fontsize=12)
plt.grid(True, linestyle="--")
plt.show()
```

> #### 2.3 Improve the Scatter Plot
- Let's add the cities' names to the chart.

```python
# same code as previous
plt.figure(figsize=(7, 5))
plt.scatter(ramadan_df["Latitude"], ramadan_df["Fasting Hours"], s=100, c="purple", alpha=0.8)
plt.title("Fasting Hours vs. Latitude", fontsize=14)
plt.xlabel("Latitude (Degrees)", fontsize=12)
plt.ylabel("Fasting Hours", fontsize=12)
plt.grid(True, linestyle="--")


# Add city names next to each dot

# .iterrows() lets us go through each row in the table one by one
# For each row, we get the data for one city (like its name, latitude, and fasting hours)
for _, row in ramadan_df.iterrows():
    # row["Latitude"] and row["Fasting Hours"] give the position of the dot
    # We add a small offset (+0.5 and +0.2) to move the text away from the dot so it's easier to see
    # row["City"] is the name we want to show next to the dot
    plt.text(row["Latitude"] + 0.5, row["Fasting Hours"] + 0.2, row["City"], fontsize=9)

    
plt.show()
```

> ### 🤔 REFLECTION: 
> Allah says, *‘He has made the sun and moon constant in their courses’* **Quran (Surah 14 : Verse 33)**. How does this explain the pattern?
